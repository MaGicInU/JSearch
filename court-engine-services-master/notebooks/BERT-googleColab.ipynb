```
from summarizer import Summarizer
model = Summarizer()
```

```

from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel

class RawText(BaseModel):
  text : str

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)

@app.get('/')
async def root():
    return {'hello': 'world'}
@app.post('/summarize')
async def summarize(textObj:RawText):
  
  result = model(textObj.text,num_sentences=15)
  return {"summary":result}
```

```
import nest_asyncio
from pyngrok import ngrok
import uvicorn

ngrok.kill()
ngrok_tunnel = ngrok.connect(8000,proto="http",options={"bind-tls":True})
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```
