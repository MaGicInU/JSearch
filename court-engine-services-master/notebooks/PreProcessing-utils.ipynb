```python
from PyPDF2 import PdfFileReader
from io import BytesIO

import re
import json
```

```python
def processPdf(fileName):
    file = open(fileName,"rb" )
    pdf = PdfFileReader(file)
    doc_text = ""
    for pageNo in range(pdf.numPages):
        page = pdf.getPage(pageNo)
        text = page.extractText()
        doc_text += text
    file.close()
    return doc_text
```
